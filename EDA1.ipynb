#Univariate Analysis(single variable)

```python
plt.figure()
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

#####insights☝️: "majority customers are between 40 to 60 age"

```python
plt.figure()
sns.histplot(df['Income'], bins=30, kde=True)
plt.title("Income Distribution")
plt.show()
```

#####insights☝️: the distribution looks fairly balnced, not extremely skewed.
#####Most data points lie between middle-income range(30000-70000)

```python
plt.figure()
sns.histplot(df['Total_Spend'], bins=30, kde=True)
plt.title("Total Spend Distribution")
plt.show()
```

#####insights☝️: the distribution has long right tail (positively skewed) which means a small number of cutomers spend very high amounts.

#Categorial Analysis

###education vs count

```python
df['Education'].value_counts().plot(kind='bar')
plt.title("Education Level Count")
plt.show()
```

###marital status distribution

```python
df['Marital_Status'].value_counts().plot(kind='bar')
plt.title("Marital Status Distribution")
plt.show
```

#Bivariate Analysis(Relation bet. two variables)

```python
plt.figure()
sns.scatterplot(x='Income', y='Total_Spend', data=df)
plt.title("Income vs Total Spend")
plt.show()
```

#####insights☝️: income and total spend shows that higher income customers generally spend more.

```python
plt.figure()
sns.scatterplot(x='Age', y='Total_Spend', data=df)
plt.title("Age vs Total Spend")
plt.show()
```

#####insights☝️: middle-aged customers (35-65) tends to spends more overall


#Profitable Product?

```python
spend_cols = [
    'MntWines','MntFruits','MntMeatProducts',
    'MntFishProducts','MntSweetProducts','MntGoldProds'
]
df[spend_cols].mean().plot(kind='bar')
plt.title("Average Spend by Product Category")
plt.show()
```

#####insights☝️: Wine is the most profitable product catogory by large margin, followed by meat products and others....

#Family Size Impact

```python
df.groupby('Family_Size')['Total_Spend'].mean().plot(kind= 'line', marker='o')
plt.title("Family Size vs Average Spend")
plt.show()
```

#####insights☝️: smaller families have the high spending, especially with size 2

#Frequency Analysis of customers

```python
plt.figure()
sns.histplot(df['Frequency'], bins=20, kde=True)
plt.title("Purcahse Frequency Distribution")
plt.show()
```

#####insights☝️: the distribution shows a dominance of repeat customers, with most of them making moderate (10-25) purchases.

#Correlation Analysis

```python
important_cols = [
    'Income',
    'Age',
    'Recency',
    'Frequency',
    'Total_Spend',
    'Family_Size',
    'Customer_Tenure_Days'
]

plt.figure(figsize=(8,6))
sns.heatmap(
    df[important_cols].corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title("Correlation Heatmap (Key Business Metrics)")
plt.show()
```

#####insights☝️: Total_Spend is strongly positively correlated with income(0.81) and purchase frequency(0.75)

#Campaign Response Analysis

```python
response_cols = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']
df[response_cols].mean().plot(kind='bar')
plt.title("Campaign Acceptance Rate")
plt.show()
```
